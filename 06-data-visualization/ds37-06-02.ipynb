# <font color='blue'> Introduction

In this lab, we'll learn about how plotting works in Pandas and Matplotlib. 

It'a important to know that Pandas uses Matplotlib behind the scenes to make plots. 

So, you will notice that Pandas plotting methods often use similar parameter names as Matplotlib methods. You can also use Matplotlib functions in combination with Pandas methods to alter the plots after drawing them. 

For example, you can use Matplotlib's `xlabel` and `title` functions to label the plot's x-axis and title, respectively, after it is drawn.

As we explore different types of plots, notice:

1. Different types of plots are drawn very similarly; they even tend to share parameter names.

2. In Pandas, calling `plot()` on a `DataFrame` is different to calling it on a `Series`. Although the methods are both named `plot`, they may take different parameters.

# <font color='blue'> Line plots
    
Recall that we can use line plots to show how a numerical variable changes over time. Let's make a line plot to show how the number of UFO sightings have changed over time.

We'll start by counting the number of UFO reports each year using ``value_counts()``, and then sort by year using ``sort_index()``. 

```python
ufo['Year'].value_counts().sort_index()
```

## Single line plots

Now we can run the ``.plot()`` method on our sorted data to generate a line plot of sightings over time. 

The result of the ``.plot()`` method is a **Pandas Axes** object; that is, an object that has all the methods and properties of a set of axes.

We can run the ``set_xlabel`` and ``set_ylabel`` methods on this set of axes to add labels to the x and y axes.

```python
ufo_axes = ufo['Year'].value_counts().sort_index().plot(title='UFO sightings over time');
ufo_axes.set_xlabel('Year')
ufo_axes.set_ylabel('Number of sightings')

```

## Multiple line plots

Let's generate a random dummy dataset to explore the ``plot`` function a bit more.

We want everyone in the class to generate the **same** random data for reproducibility purposes. We tell Numpy to do this by setting a **seed**; as long as the seed is initialised with the same number between two different people, Numpy will generate the same numbers each time.

```python
np.random.seed(42) #

df = pd.DataFrame(np.random.randn(10, 4), 
                  columns=['col1', 'col2', 'col3', 'col4'],
                  index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
```

We've generated four time series. It's easy to call the `plot` method on a DataFrame containing multiple time series, and have them plotted in one go.

```python
df.head()
```

```python
df.plot();
```

## Changing line styles

It's easy to customise the appearance of lines.

Let's plot both `col1` and `col4` of `df` against the index on the same axes. 

We can use the `style` parameter to change the appearance of both plots to match the example below. 

`:` denotes dotted line, `v` denotes triangle marker, `r` denotes red and `b` blue. The `figsize` is 15 by 7.

```python
# : - dotted line, v - triangle_down
# r - red, b - blue
df[['col1', 'col4']].plot(figsize=(15,7), style={'col1': 'r:', 'col4': 'vb'});
```

## Simple bar charts

Now we want to make a visualisation that ranks the continents according to the number of countries in each continent. 

Since we're **ranking** quantities, a bar chart is a good fit.

We start by getting the counts by continent in our DataFrame. 

```python
drinks.groupby('continent').count()['country']
```

## Grouped bar charts

Let's use `groupby` and `mean` to calculate the mean alcohol consumption per continent.

```python
drinks.groupby('continent').mean()
```

Now let's visualise this breakdown by continent using a bar chart. 

```python
# Side-by-side bar plots
drinks.groupby('continent').mean().plot(kind='bar',title='Mean alcohol consumption by continent');
```

We can order the `continent` x-axis by the `beer_servings` column and re-generate the plot.

```python
drinks.groupby('continent').mean().sort_values('beer_servings').plot(kind='bar',title='Mean alcohol consumption by continent');
```

# <font color='blue'> Box plots

We can use boxplots to quickly summarize distributions.

**Five-number summary:**

- min = minimum value
- 25% = first quartile (Q1) = median of the lower half of the data
- 50% = second quartile (Q2) = median of the data
- 75% = third quartile (Q3) = median of the upper half of the data
- max = maximum value

(It's more useful than mean and standard deviation for describing skewed distributions.)

**Interquartile Range (IQR)** = Q3 - Q1

**Outliers:**

- below Q1 - 1.5 * IQR
- above Q3 + 1.5 * IQR

```python
df.boxplot();
```

Let's see how box plots are generated so we can best interpret them.

```python
# Show "five-number summary" for spirit.
drinks['spirit_servings'].describe()
```

```python
# Compare with box plot.
drinks['spirit_servings'].plot(kind='box');
```

```python
drinks.boxplot(column='beer_servings', by='continent');
```

# <font color='blue'> Scatter plots
    
Scatter plots can be used to show the relationship between two variables. 

Let's do this using the drinks dataset, to show the relationship between beer consumption and wine consumption.

```python
drinks.sort_values(by='beer_servings').plot(kind='scatter',x='beer_servings',y='wine_servings')

```

We can change the transparency of the dots to 0.3 using the `alpha` parameter.

```python
drinks.sort_values(by='beer_servings').plot(kind='scatter',x='beer_servings',y='wine_servings',alpha=0.3)
```

Vary the colour of each point according to the value of the `spirit_serving` column, using the `c` argument and the `colormap` option.


```python
drinks.sort_values(by='beer_servings').plot(kind='scatter',x='beer_servings',y='wine_servings',c='spirit_servings',colormap='Blues');

```
